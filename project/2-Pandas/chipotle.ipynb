## Basic Level

### Part 1: Read in the file with `csv.reader()` and store it in an object called `file_nested_list`.

Hint: This is a TSV (tab-separated value) file, and `csv.reader()` needs to be told [how to handle it](https://docs.python.org/2/library/csv.html).

```python
import csv
from collections import namedtuple   # Convenient to store the data rows

DATA_FILE = './data/chipotle.tsv'
```

```python
file_nested_list=[]
with open(DATA_FILE) as tsv_file:
    tsv_reader = csv.reader(tsv_file, delimiter='\t')
    for row in tsv_reader:
        file_nested_list.append(row)
file_nested_list
```

### Part 2: Separate `file_nested_list` into the `header` and the `data`.


```python
header=file_nested_list[0]
header
```

```python
data=file_nested_list[1:]
data
```

## Advanced Level


### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
#df[df.item_name.str.contains('Burrito')]
df[df.item_name.str.contains('Burrito')].sort_values('item_name',ascending=False)
```

```python
#using python regex, count the number of ',' and add 1 to get total number of toppings. x,y => 2 toppings
import re
df['Toppings'] = df[df.item_name.str.contains('Burrito')].choice_description.apply(lambda x: len(re.findall(',',str(x)))+1)
df[df.item_name.str.contains('Burrito')].Toppings.mean()
```
